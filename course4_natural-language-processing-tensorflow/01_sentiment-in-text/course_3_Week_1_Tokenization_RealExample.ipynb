```python
english_stopwords_list = list(en_stop)
len(english_stopwords_list)
```

```python
def prepare_english_text(path):
    sentences = []
    labels = []
    
    english_stopwords_list = list(en_stop)
    print(f"length of En stopwords is: {len(english_stopwords_list)}")
    
    with open(path,'r') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        next(reader)
        
        for row in reader:
            labels.append(row[0])
            sentence = row[1]
            
            for word in english_stopwords_list:
                token = " " + word + " "
                sentence = sentence.replace(token, " ")
            sentences.append(sentence)
        print(f"Length of sentences, labels is {len(sentences)},{len(labels)}")   
        return sentence, labels
```

```python
sentences, labels = prepare_english_text("/tmp/bbc-text.csv")
```

```python
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
print(len(word_index))
```

```python
sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences, padding='post')
print(padded[0])
print(padded.shape)
```

```python
label_tokenizer = Tokenizer()
label_tokenizer.fit_on_texts(labels)
label_word_index = label_tokenizer.word_index
label_seq = label_tokenizer.texts_to_sequences(labels)
print(label_seq)
print(label_word_index)
```
